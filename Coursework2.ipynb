```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
cdc_diabetes_health_indicators = fetch_ucirepo(id=891) 
  
# data (as pandas dataframes) 
X = cdc_diabetes_health_indicators.data.features 
y = cdc_diabetes_health_indicators.data.targets 
  
# Combining features and target into one DataFrame
data = pd.concat([X, y], axis=1)
```

```python
# Research Questions
# 1. How does the prevalence of diabetes vary across different age groups

# Define age category labels
age_labels = {
    1: '18-24', 2: '25-29', 3: '30-34', 4: '35-39',
    5: '40-44', 6: '45-49', 7: '50-54', 8: '55-59',
    9: '60-64', 10: '65-69', 11: '70-74', 12: '75-79', 13: '80 or older'
}

# Map integer codes to labels
data['Age'] = data['Age'].map(age_labels)

# Group data by Age and calculate the mean of Diabetes_binary to get the proportion
age_diabetes = data.groupby('Age')['Diabetes_binary'].mean()

# Plotting
plt.figure(figsize=(10, 6))
age_diabetes.sort_index(key=lambda x: x.map({v: k for k, v in age_labels.items()})).plot(kind='bar', color='skyblue')
plt.title('Prevalence of Diabetes by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Proportion with Diabetes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the number of occurrences for each age group and diabetes status
age_diabetes_count = data.groupby(['Age', 'Diabetes_binary']).size().unstack()

# Plotting the counts of diabetes status ('0' and '1') for each age group
age_diabetes_count.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'])
plt.title('Count of Diabetes Status by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Diabetes Status', labels=['No Diabetes (0)', 'Prediabetes/Diabetes (1)'])
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

## Conclusion:

* Increasing Trend: There appears to be a clear trend that as age increases, the proportion of individuals with diabetes also increases. This is expected as the risk of diabetes typically rises with age.

* Lowest Prevalence in Youth: The youngest age groups, starting from 18-24 up to 35-39, have a relatively low prevalence of diabetes. This could be due to a combination of factors such as higher metabolic rates, more physical activity, and fewer age-related metabolic changes.

* Significant Increase Middle-Aged Onwards: From the 40-44 age group and onwards, there is a noticeable increase in the prevalence of diabetes. This could indicate that middle age is a significant time for the onset or diagnosis of diabetes.

* Highest Prevalence in Seniors: The highest proportions of diabetes are seen in the older age brackets, 70-74, 75-79, and 80 or older. This may reflect the cumulative effect of lifestyle, dietary habits, and other health conditions that escalate the risk of diabetes.

* Possible Plateau or Decrease After 75-79: Interestingly, the proportion of individuals with diabetes in the "80 or older" category is slightly lower than the 75-79 group. This might be due to survivorship bias where healthier seniors are more likely to reach this age, or it might be an artifact of the data.

Overall, the graph suggests that age is a significant factor in the prevalence of diabetes, with increased age correlating with higher rates of the condition. 

```python
# 2. Is there a relationship between body mass index and the likelihood of diabetes?

# Calculate mean BMI for each diabetes status
mean_bmi_by_diabetes = data.groupby('Diabetes_binary')['BMI'].mean()

# Plotting a bar graph
plt.figure(figsize=(10, 6))
mean_bmi_by_diabetes.plot(kind='bar', color=['blue', 'orange'])

plt.title('Average BMI by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Average BMI')
plt.xticks(ticks=[0, 1], labels=['No Diabetes (0)', 'Prediabetes/Diabetes (1)'], rotation=0)

plt.show()
```

```python
# Setting the bins for BMI values
bins = range(10, 100, 2)

# Create the stacked histogram
plt.figure(figsize=(12, 8))

# Histogram for individuals without diabetes
plt.hist(data['BMI'][data['Diabetes_binary'] == 0], bins=bins, alpha=0.5, label='No Diabetes (0)', color='blue', edgecolor='black')

# Histogram for individuals with prediabetes or diabetes, stacked on top of the previous histogram
plt.hist(data['BMI'][data['Diabetes_binary'] == 1], bins=bins, alpha=0.5, label='Prediabetes/Diabetes (1)', color='orange', edgecolor='black')

# Adding labels and title
plt.title('Stacked Histogram of BMI by Diabetes Status')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.legend()

# Show the plot
plt.show()
```

### Conclusions:

* Higher Average BMI in Diabetic Patients: The average BMI for individuals with prediabetes or diabetes is higher than for those without diabetes. This supports the commonly understood correlation where a higher BMI is associated with an increased risk of developing diabetes.

* BMI as a Risk Factor: The graph underscores the role of BMI as a potential risk factor for diabetes. Medical literature suggests that a higher body mass index can lead to insulin resistance, which is a key factor in the development of type 2 diabetes.

* Common BMI Range: The most common range of BMI for both groups appears to be around the same interval, which seems to be in the overweight range (considering general BMI classifications). This suggests that a high BMI is common among both those with and without diabetes in your dataset.

* Presence of Diabetic Individuals in All BMI Ranges: The orange segments, indicating individuals with prediabetes or diabetes, are present across all BMI ranges. This presence across the spectrum emphasizes that diabetes can occur at any BMI, though it may be more prevalent at higher BMI levels.

* Diabetes Across BMI Categories: The presence of the orange bars within most BMI intervals indicates that diabetes affects individuals across various body weights, reinforcing that while higher BMI is a risk factor, it's not the only one.

```python
# 3. What is the impact of smoking on the general health of individuals with diabetes?

# Filter data for diabetics only
diabetic_patients = data[data['Diabetes_binary'] == 1]

# Calculate the mean general health rating for smokers and non-smokers within diabetic patients
gen_health_by_smoker_status = diabetic_patients.groupby('Smoker')['GenHlth'].mean().reset_index()

# Grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35 

# Plotting the bars for each smoker status
ax.bar(gen_health_by_smoker_status['Smoker'] - width/2, gen_health_by_smoker_status['GenHlth'], width, label='Non-Smokers', color='green')
ax.bar(gen_health_by_smoker_status['Smoker'] + width/2, gen_health_by_smoker_status['GenHlth'], width, label='Smokers', color='red')

ax.set_title('General Health Ratings of Diabetic Patients by Smoking Status')
ax.set_xlabel('Smoking Status')
ax.set_ylabel('Average General Health Rating')
ax.set_xticks(gen_health_by_smoker_status['Smoker'])
ax.set_xticklabels(['Non-Smokers (0)', 'Smokers (1)'])
ax.legend()

plt.show()
```

```python
# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Smoker', y='GenHlth', data=diabetic_patients, palette="Set3")

# Set the title and labels
plt.title('Box Plot of General Health Ratings for Diabetic Patients by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('General Health Rating')
plt.xticks([0, 1], ['Non-Smokers (0)', 'Smokers (1)'])

# Show the plot
plt.show()
```

### Conclusions

* Close Average Ratings: The general health ratings appear similar for both non-smokers and smokers, indicated by the bars' comparable heights. This might suggest that, within this population of diabetic individuals, smoking status is not substantially differentiating their self-reported general health.

* Median Health Ratings: Both non-smokers and smokers have median health ratings that indicate neither group reports optimal health (since neither median is at '1' for excellent health), which is consistent with them having diabetes.

* Presence of Outliers: There are outliers in both groups, indicating that there are individuals who perceive their health as better than the majority of their respective group (ratings closer to 1).

* The box plot reveals that smoking status might not be a significant differentiator in perceived health among these diabetic patients, or that the effects of diabetes on perceived health might be more significant than the effects of smoking

```python
# 4. How does physical activity influence the physical and mental health of diabetic patients?

average_scores = diabetic_patients.groupby('PhysActivity').agg({
    'MentHlth': 'mean',
    'PhysHlth': 'mean'
}).reset_index()

# Plotting
fig, ax = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Mental Health scores line graph
ax[0].plot(average_scores['PhysActivity'], average_scores['MentHlth'], label='Mental Health', marker='o')
ax[0].set_title('Average Mental Health Score by Physical Activity Status')
ax[0].set_ylabel('Average Mental Health Score')
ax[0].legend()

# Physical Health scores line graph
ax[1].plot(average_scores['PhysActivity'], average_scores['PhysHlth'], label='Physical Health', color='orange', marker='o')
ax[1].set_title('Average Physical Health Score by Physical Activity Status')
ax[1].set_xlabel('Physical Activity Status')
ax[1].set_ylabel('Average Physical Health Score')
ax[1].legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the mean scores for mental and physical health for active and inactive patients
average_scores = diabetic_patients.groupby('PhysActivity')[['MentHlth', 'PhysHlth']].mean()

# Create the stacked area chart
plt.figure(figsize=(10, 6))

# Define the x-axis labels
labels = ['Inactive (0)', 'Active (1)']

# Plotting the areas
plt.fill_between(labels, average_scores['MentHlth'], color='skyblue', label='Average Mental Health Score')
plt.fill_between(labels, average_scores['PhysHlth'], color='orange', label='Average Physical Health Score', alpha=0.5)

# Adding labels and title
plt.title('Stacked Area Chart of Health Scores by Physical Activity Status')
plt.xlabel('Physical Activity Status')
plt.ylabel('Average Health Score')
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

### Conclusions

* Trend of Health Scores: Both graphs indicate a downward trend, suggesting that as physical activity status increases (moving from inactive to active), both mental and physical health scores improve (since lower scores mean better health).

* Stronger Correlation with Physical Health: The decline in average physical health score with increasing physical activity is more pronounced than the mental health score. This could suggest that physical activity has a more direct impact on physical health for diabetic patients.

* Consistent Impact: Physical activity appears to have a consistent positive impact on both mental and physical health, highlighting its importance in the management of diabetes.

* Physical Activity Association: Active individuals (1 on the chart) tend to report better average mental and physical health scores (lower scores indicate better health) than inactive individuals (0 on the chart). This is evident from the decrease in the area representing health scores as we move from 'Inactive' to 'Active'.

* Overlap in Scores: There's an overlap between mental and physical health scores, highlighting that individuals' perceptions of their mental and physical health are not isolated; they influence each other.

```python
# 5. Does high blood pressure co-occur with diabetes, and how is this related to age and sex?

# Map the binary 'Sex' values to 'Female' and 'Male'
data['Sex'] = data['Sex'].map({0: 'Female', 1: 'Male'})

# Create the 'HighBP_Diabetes' column which is True if both 'HighBP' and 'Diabetes_binary' are 1
data['HighBP_Diabetes'] = (data['HighBP'] == 1) & (data['Diabetes_binary'] == 1)

# Group by 'Age' and 'Sex' and calculate the mean for 'HighBP_Diabetes'
co_occurrence = data.groupby(['Age', 'Sex'])['HighBP_Diabetes'].mean().unstack()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(co_occurrence, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Co-occurrence of High Blood Pressure and Diabetes by Age Group and Sex')
plt.ylabel('Age Group')
plt.xlabel('Sex')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### Conclusions

* Age-Related Increase: There's a clear trend that shows the co-occurrence of high blood pressure and diabetes increases with age. This is seen in the gradient of color becoming darker as age increases, which is consistent with the understanding that the risk of both conditions increases with age.

* Sex Differences: In almost all age groups, males exhibit a higher co-occurrence rate than females. This suggests that in this dataset, males with diabetes have a higher likelihood of also having high blood pressure compared to their female counterparts.

* Peak in Older Age Groups: The co-occurrence rate seems to peak in the '70-74' age group for both sexes, with a particularly high rate for males. After this peak, there's a slight decrease in the oldest age group ('80 or older').

* Younger Age Groups: For the younger age groups ('18-24' up to '45-49'), the co-occurrence rates are relatively low, although they are not zero, indicating that these conditions can and do appear in younger individuals as well.

* Middle Age Onset: The co-occurrence rates start to become more pronounced from the '50-54' age group onwards, highlighting the middle age as a critical period for the onset or diagnosis of these conditions.

* These observations underscore the importance of monitoring blood pressure in diabetic patients, especially as they age and if they are male. It also highlights the potential for targeted interventions in middle-aged populations to prevent or manage these conditions.

```python
# 6. What is the role of diet (consumption of fruits and vegetables) in managing diabetes?

fruit_consumption = data.groupby('Diabetes_binary')['Fruits'].value_counts(normalize=True).unstack()
veggie_consumption = data.groupby('Diabetes_binary')['Veggies'].value_counts(normalize=True).unstack()

# Create donut charts
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Donut chart for fruit consumption
ax[0].pie(fruit_consumption[1], labels=['No Diabetes', 'Diabetes'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
ax[0].add_artist(plt.Circle((0,0),0.70,fc='white'))
ax[0].set_title('Fruit Consumption')

# Donut chart for vegetable consumption
ax[1].pie(veggie_consumption[1], labels=['No Diabetes', 'Diabetes'], autopct='%1.1f%%', startangle=90, colors=['#99ff99','#ffcc99'])
ax[1].add_artist(plt.Circle((0,0),0.70,fc='white'))
ax[1].set_title('Vegetable Consumption')

# Display the plot
plt.tight_layout()
plt.show()
```

### Conclusions

* The chart shows that 52.3% of the surveyed population who do not have diabetes consume fruit.
Conversely, 47.7% of the surveyed population with diabetes consume fruit.

* Fruit consumption is almost equally distributed between those with diabetes and those without. The difference is quite small, suggesting that having diabetes does not significantly change the likelihood of consuming fruit in this population.

* The chart indicates that 52% of the surveyed population without diabetes consume vegetables.
Meanwhile, 48% of the surveyed population with diabetes consume vegetables.

* Vegetable consumption shows a similar pattern to fruit consumption, with a very slight majority of vegetable consumption occurring in individuals without diabetes.

* The difference in consumption rates for both fruits and vegetables between those with and without diabetes is marginal (within 4-5%), which could indicate that a diabetes diagnosis does not drastically alter consumption habits regarding fruits and vegetables.

*  it might be suggested that both groups—those with and without diabetes—are consuming fruits and vegetables at similar rates, which may highlight a positive trend in dietary choices among people with diabetes


```python
# 7. What is the association between alcohol consumption and diabetes status?

# Calculate the proportions of heavy alcohol consumption for non-diabetics
no_diabetes_counts = data[data['Diabetes_binary'] == 0]['HvyAlcoholConsump'].value_counts(normalize=True)

# Calculate the proportions of heavy alcohol consumption for diabetics
with_diabetes_counts = data[data['Diabetes_binary'] == 1]['HvyAlcoholConsump'].value_counts(normalize=True)

# Set up the figure for two side-by-side pie charts
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for non-diabetics
axs[0].pie(no_diabetes_counts, labels=['Non-Heavy Drinkers', 'Heavy Drinkers'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
axs[0].set_title('Non-Diabetics')

# Pie chart for diabetics
axs[1].pie(with_diabetes_counts, labels=['Non-Heavy Drinkers', 'Heavy Drinkers'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'orange'])
axs[1].set_title('Diabetics')

# Set the overall title
plt.suptitle('Proportion of Heavy Alcohol Consumption by Diabetes Status')

# Show the plot
plt.show()
```

### Conclusions

* Lower Proportion Among Diabetics: The pie chart shows that a smaller proportion of individuals with diabetes are heavy drinkers compared to those without diabetes (2.4% vs. 6.1%). This could indicate that those with diabetes may be more cautious about their alcohol consumption due to the interaction with their condition or medication.

* Majority Non-Heavy Drinkers: In both groups, the vast majority are not heavy drinkers. This is positive from a health perspective, as heavy drinking can exacerbate diabetes complications.

* Health Awareness or Reporting Bias: The lower proportion of heavy drinkers among diabetics could be due to increased health awareness or possibly a reporting bias if diabetics are less likely to report heavy drinking.


```python
# 8. Does the incidence of stroke or heart disease differ between diabetic and non diabetic individuals? 

# Calculate the proportion of stroke and heart disease in diabetic and non-diabetic groups
stroke_proportions = data.groupby('Diabetes_binary')['Stroke'].mean()
heart_disease_proportions = data.groupby('Diabetes_binary')['HeartDiseaseorAttack'].mean()

# Prepare the figure
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# Donut chart for Stroke
axs[0].pie(stroke_proportions, labels=['No Stroke', 'Stroke'], autopct='%1.1f%%', startangle=90, colors=['lightgray', 'skyblue'], wedgeprops=dict(width=0.4))
axs[0].set_title('Proportion of Stroke by Diabetes Status')

# Donut chart for Heart Disease or Attack
axs[1].pie(heart_disease_proportions, labels=['No Heart Disease/Attack', 'Heart Disease/Attack'], autopct='%1.1f%%', startangle=90, colors=['lightgray', 'lightgreen'], wedgeprops=dict(width=0.4))
axs[1].set_title('Proportion of Heart Disease or Attack by Diabetes Status')

# Set the overall title
plt.suptitle('Proportion of Stroke and Heart Disease by Diabetes Status')

# Show the plot
plt.tight_layout()
plt.show()
```

### Conclusions

* Stroke Proportion: There's a significant difference in the occurrence of stroke between diabetic and non-diabetic individuals, with diabetics having a higher reported rate of stroke.

* Heart Disease/Attack Proportion: Similarly, the occurrence of heart disease or heart attacks is higher in diabetic individuals compared to non-diabetics.

* Comparative Risk: The increased proportions in both stroke and heart disease among diabetic patients compared to non-diabetics suggest a stronger association of these conditions with diabetes. This emphasizes the cardiovascular risks associated with diabetes.

* Health Management Implications: The data implies that monitoring and managing cardiovascular health is crucial for diabetic patients.
